## Applied Macroeconometrics 1 Term Paper data preparation.

#### Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

#### Load Francis and Ramey (2009) data

```python
Francis_Ramey = pd.read_excel("Data/Francis-Ramey_JMCB_Data_09.xls", sheet_name="Data for SVARs",header=3, index_col=0)
```

```python
Francis_Ramey.head()
```

#### Load data for Current tax receipts (National Data Table 3.2)

```python
taxes = pd.read_csv("Data/taxes.csv", header=3, index_col=1)
taxes = taxes.iloc[1:]
taxes = taxes.drop(taxes.columns[0], axis=1)
taxes = taxes.iloc[:, 4:244]
```

```python
taxes.head()
```

#### Load data for gdp deflator (National Data Table Table 1.1.4)

```python
gdp_deflator = pd.read_csv("Data/GDP_deflator.csv", header=3, index_col=1)
gdp_deflator = gdp_deflator.iloc[1:]
gdp_deflator = gdp_deflator.drop(gdp_deflator.columns[0], axis=1)
gdp_deflator = gdp_deflator.iloc[:, 1:61]
```

```python
gdp_deflator.head()
```

#### Select average hours worked and real GDP from Francis Ramey dataset.

```python
data = Francis_Ramey[["avgwkhrs_adj", "rgdp"]]

data["Current tax receipts"] = taxes.iloc[1].values

data.reset_index(inplace=True)
```

```python
# Extract querter for gdp deflator calculations. 
def extract_quarter(decimal_value):
    year = int(decimal_value)  # Extract the integer part as the year
    remainder = decimal_value - year  # Get the decimal part to determine the quarter
    
    if remainder == 0.0:
        quarter = 'Q1'
    elif remainder == 0.25:
        quarter = 'Q2'
    elif remainder == 0.5:
        quarter = 'Q3'
    elif remainder == 0.75:
        quarter = 'Q4'
    else:
        quarter = 'Unknown'
    
    return f"{quarter}-{year}"

# Create a new column with the quarter-year information
data["Quarter Name"] = data["quarter"].apply(extract_quarter)
data["Year"] = data["Quarter Name"].str.split("-").str[-1].astype(int)

gdp_transposed = gdp_deflator.T

# Create a mapping for yearly GDP
gdp_mapping = dict(zip(gdp_transposed.index, gdp_transposed["        Gross domestic product"]))
gdp_mapping= {int(k): pd.to_numeric(v, errors='coerce') for k, v in gdp_mapping.items()}

data = data.astype({"Year": float, "Current tax receipts" : float})
```

#### Create columns for Tax revenues and working age population. 

```python
data["Tax revenues"] = data.apply(lambda row: row["Current tax receipts"] / gdp_mapping[row["Year"]], axis=1)

data["population aged 16+"] = Francis_Ramey["civnipop16"].values

data["Tax revenues"] = data["Tax revenues"] / data["population aged 16+"]
```

#### Plot data

```python
plt.plot(data["quarter"],data["avgwkhrs_adj"] )
```

```python
plt.plot(data["quarter"],data["Tax revenues"] )
```

```python
plt.plot(data["quarter"],np.log(data["rgdp"]) )
```

#### Get productivity data

```python
productivity = pd.read_excel("Francis-Ramey_JMCB_Data_09.xls", sheet_name="Productivity_adj",header=9, index_col=0)
```

```python
data["labor productivity"] = productivity["x"].values
```

```python
plt.plot(data["quarter"],np.log(data["labor productivity"]))
```
